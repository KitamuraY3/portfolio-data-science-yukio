# 1. BIBLIOTECAS:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# 2. DATASET:

```python
# CARREGAR O DATASET 'Iris' DO SCIKIT-LEARN:
from sklearn.datasets import load_iris

# DATASET:
iris_raw = load_iris(as_frame=True)
iris = iris_raw.frame

# TRANSFORMAR O TARGET DE C√ìDIGO PARA O NOME REAL DAS ESP√âCIES:
iris['species'] = iris_raw.target_names[iris['target']]

# DROP DA COLUNA'target' (C√ìDIGO NUM√âRICO), AGORA TEMOS A 'species':
iris.drop(columns='target', inplace=True)
```

```python
iris
```

### üîç Estrutura do dataset Iris

O dataset cont√©m 150 amostras de flores, divididas igualmente entre 3 esp√©cies:

- **setosa**
- **versicolor**
- **virginica**

Cada flor √© descrita por 4 vari√°veis num√©ricas:

- Comprimento da s√©pala (`sepal length (cm)`)
- Largura da s√©pala (`sepal width (cm)`)
- Comprimento da p√©tala (`petal length (cm)`)
- Largura da p√©tala (`petal width (cm)`)

# 3. AN√ÅLISE DESCRITIVA INICIAL:

```python
# DESCRIBE:
iris.describe()
```

### üìà Resumo Estat√≠stico ‚Äì Vari√°veis Num√©ricas

A tabela acima mostra que:

- O comprimento da p√©tala (`petal length`) apresenta grande varia√ß√£o, com m√©dia de aproximadamente 3,76 cm e desvio padr√£o comparativamente elevado de 1,77 cm.
- J√° a largura da s√©pala (`sepal width`) tem menor dispers√£o (desvio padr√£o de 0,44 cm) e distribui√ß√£o mais concentrada.
- Algumas vari√°veis, como `sepal width`, podem conter outliers (diferen√ßa grande entre `min` e `25%`).


# 4. VISUALIZA√á√ïES UNIVARIADAS COM GR√ÅFICOS:

## 4.1. HISTOGRAMAS:

### 4.1.1. HISTOGRAMA 'sepal length':

```python
plt.figure(figsize=(8, 5))
plt.grid(True, linestyle='-', alpha=0.5)
sns.histplot(data=iris, x='sepal length (cm)', bins=20, kde=True)

plt.title('Distribui√ß√£o do Comprimento da S√©pala', fontsize=14)
plt.xlabel('Comprimento da S√©pala (cm)')
plt.ylabel('Frequ√™ncia')
plt.tight_layout()
plt.show()
```

#### üìå Interpreta√ß√£o: `sepal length (cm)`

- A distribui√ß√£o do comprimento da s√©pala √© aproximadamente sim√©trica, com concentra√ß√£o entre 5,0 e 6,5 cm.  
- A curva de densidade indica que n√£o h√° grandes picos ou caudas pronunciadas, sugerindo uma distribui√ß√£o pr√≥xima da normal.  
- N√£o h√° evid√™ncias claras de outliers extremos nesta vari√°vel.

### 4.1.2. HISTOGRAMA 'sepal width':

```python
plt.figure(figsize=(8, 5))
plt.grid(True, linestyle='-', alpha=0.5)
sns.histplot(data=iris, x='sepal width (cm)', bins=20, kde=True)

plt.title('Distribui√ß√£o da Largura da S√©pala', fontsize=14)
plt.xlabel('Largura da S√©pala (cm)')
plt.ylabel('Frequ√™ncia')
plt.tight_layout()
plt.show()
```

#### üìå Interpreta√ß√£o: `sepal width (cm)`

- A distribui√ß√£o da largura da s√©pala mostra uma leve assimetria, com a maioria dos valores concentrados entre 2,8 e 3,4 cm.

- √â poss√≠vel notar um n√∫mero pequeno de observa√ß√µes com valores abaixo de 2,5 cm, que podem representar outliers. A curva de densidade tamb√©m sugere um leve achatamento da distribui√ß√£o, indicando uma dispers√£o mais ampla dos dados.


### 4.1.3. HISTOGRAMA 'petal lenght':

```python
plt.figure(figsize=(8, 5))
plt.grid(True, linestyle='-', alpha=0.5)
sns.histplot(data=iris, x='petal length (cm)', bins=20, kde=True)

plt.title('Distribui√ß√£o do Comprimento da P√©tala', fontsize=14)
plt.xlabel('Comprimento da P√©tala (cm)')
plt.ylabel('Frequ√™ncia')
plt.tight_layout()
plt.show()
```

#### üìå Interpreta√ß√£o: `petal length (cm)`

A vari√°vel `petal length (cm)` apresenta uma distribui√ß√£o **bimodal clara**, com dois grupos distintos de valores:

- Um grupo com comprimentos entre 1 e 2;
- Outro entre 4 e 6 cm.

Essa separa√ß√£o sugere que o comprimento da p√©tala √© uma vari√°vel altamente discriminativa entre as esp√©cies de flores, sendo potencialmente √∫til em tarefas de classifica√ß√£o.

### 4.1.4. HISTOGRAMA 'petal width':

```python
plt.figure(figsize=(8, 5))
plt.grid(True, linestyle='-', alpha=0.5)
sns.histplot(data=iris, x='petal width (cm)', bins=20, kde=True)

plt.title('Distribui√ß√£o da Largura da P√©tala', fontsize=14)
plt.xlabel('Largura da P√©tala (cm)')
plt.ylabel('Frequ√™ncia')
plt.tight_layout()
plt.show()
```

#### üìå Interpreta√ß√£o: `petal width (cm)`

A largura da p√©tala apresenta uma distribui√ß√£o com **tr√™s grupos visuais bem definidos**, o que indica grande potencial dessa vari√°vel para diferenciar esp√©cies. Os tr√™s grupos s√£o:

- Valores pr√≥ximos de 0,2 cm;
- Valores intermedi√°rios de 1,0 a 1,5 cm;
- Valores acima de 1,5 cm.

Essa distribui√ß√£o refor√ßa que `petal width` pode ser um dos atributos mais relevantes para modelos de classifica√ß√£o com esse conjunto de dados.

## 4.2. BOXPLOT:

### 4.2.1. BOXPLOT 'sepal lenght' POR 'species':

```python
plt.figure(figsize=(8, 5))
plt.grid(True, linestyle='-', alpha=0.5)
sns.boxplot(data=iris, x='species', y='sepal length (cm)')

plt.title('Comprimento da S√©pala por Esp√©cie', fontsize=14)
plt.xlabel('Esp√©cie')
plt.ylabel('Comprimento da S√©pala (cm)')
plt.tight_layout()
plt.show()
```

#### üìå Interpreta√ß√£o: `sepal length (cm)` por esp√©cie

- A esp√©cie *Setosa* possui s√©palas visivelmente mais curtas, com baixa varia√ß√£o.  
- A esp√©cie *Virginica* tem os maiores valores de comprimento da s√©pala e tamb√©m maior dispers√£o.
- A esp√©cie *Versicolor* tem valores medianos de comprimento da s√©pala com dispers√£o grande.

H√° sobreposi√ß√£o parcial entre as distribui√ß√µes de *Versicolor* e *Virginica*, o que pode dificultar a separa√ß√£o entre essas duas esp√©cies com base apenas nessa vari√°vel.


### 4.2.2. BOXPLOT DE 'sepal wigth' POR 'species':

```python
plt.figure(figsize=(8, 5))
plt.grid(True, linestyle='-', alpha=0.5)
sns.boxplot(data=iris, x='species', y='sepal width (cm)')

plt.title('Largura da S√©pala por Esp√©cie', fontsize=14)
plt.xlabel('Esp√©cie')
plt.ylabel('Largura da S√©pala (cm)')
plt.tight_layout()
plt.show()
```

#### üìå Interpreta√ß√£o: `sepal width (cm)` por esp√©cie

- A esp√©cie *Setosa* apresenta a maior mediana de largura da s√©pala entre as tr√™s classes, al√©m de uma distribui√ß√£o mais concentrada e sim√©trica.

- As esp√©cies *Versicolor* e *Virginica* possuem distribui√ß√µes semelhantes, com *Virginica* apresentando maior dispers√£o e alguns poss√≠veis outliers com valores baixos.

Essa vari√°vel mostra algum poder discriminativo, mas com menos clareza que outras como `petal length`.

### 4.2.3. BOXPLOT DE 'petal lenght' POR 'species':

```python
plt.figure(figsize=(8, 5))
plt.grid(True, linestyle='-', alpha=0.5)
sns.boxplot(data=iris, x='species', y='petal length (cm)')

plt.title('Comprimento da P√©tala por Esp√©cie', fontsize=14)
plt.xlabel('Esp√©cie')
plt.ylabel('Comprimento da P√©tala (cm)')
plt.tight_layout()
plt.show()
```

#### üìå Interpreta√ß√£o: `petal length (cm)` por esp√©cie

O comprimento da p√©tala √© a vari√°vel com **maior poder discriminativo** entre as esp√©cies:

- *Setosa* possui p√©talas curtas, com valores concentrados abaixo de 2 cm;
- *Versicolor* ocupa uma faixa intermedi√°ria, entre 3 e 5 cm;
- *Virginica* apresenta os maiores valores, geralmente acima de 5 cm.

As distribui√ß√µes n√£o se sobrep√µem, tornando essa vari√°vel altamente eficaz para a separa√ß√£o entre classes em um modelo de classifica√ß√£o.

### 4.2.4. BOXPLOT DE 'petal width' POR 'species':

```python
plt.figure(figsize=(8, 5))
plt.grid(True, linestyle='-', alpha=0.5)
sns.boxplot(data=iris, x='species', y='petal width (cm)')

plt.title('Largura da P√©tala por Esp√©cie', fontsize=14)
plt.xlabel('Esp√©cie')
plt.ylabel('Largura da P√©tala (cm)')
plt.tight_layout()
plt.show()
```

#### üìå Interpreta√ß√£o: `petal width (cm)` por esp√©cie

A largura da p√©tala tamb√©m mostra alto poder discriminativo:

Apesar de uma leve sobreposi√ß√£o entre *Versicolor* e *Virginica*, essa vari√°vel continua sendo uma excelente candidata para separar classes em uma an√°lise supervisionada.
